# 0 - Data Set Merging and Preparation - Lecture

```python
import pandas as pd
lecture_events = pd.read_csv('events_pi1_vorlesung.csv', delimiter=";")
```

```python
small_events = lecture_events
del small_events['duration']
del small_events['r_id']
del small_events['resource_id_1']
del small_events['course_id']
small_events.head()
```

```python
lecture_resources = pd.read_csv('pi1_lecture_resources.csv', delimiter=";")
```

```python
small_resources = lecture_resources
del small_resources['externalId']
del small_resources['lastUpdate']
del small_resources['parentId']
small_resources.head()
```

```python
result = pd.merge(small_events, small_resources, how='left', left_on = 'resource_id', right_on = 'leapId')
result.head()
```

```python
result.to_csv('events-and-resources.csv', sep=";", encoding="utf-8")
```
